Importing useful modules

```python
from __future__ import division # in the future (i.e. Python3), all division favors mathematicians
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

Sample 1,000 values from standard normal, 10,000 times

```python
for n, b in zip([100, 500, 1000, 10000], [10, 30, 80, 100]):
    X = np.random.randn(n, 1000)
    S = sum(X, axis=1) / n
    plt.hist(S , bins=b)
    plt.show()
    print(n, np.mean(S), np.var(S))
```
