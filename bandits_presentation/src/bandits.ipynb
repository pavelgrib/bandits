```python
from __future__ import division # divide and conquer
import numpy as np
import scipy as sp
from scipy.stats import stats
import pandas as pd
from operator import itemgetter
import matplotlib.pyplot as plt
```

```python
class Bandit(object):
    def __init__(self, actual_perf):
        self.alpha = 1
        self.beta = 1
        self.actual = actual_perf
    
    def sample(self):
        return np.random.beta(self.alpha, self.beta)

    def update(self, randnum):
        if randnum < self.actual:
            self.alpha += 1
        else:
            self.beta += 1
```

```python
perfs = [0.01, 0.008, 0.022, 0.019, 0.038, 0.036, 0.029, 0.015, 0.003, 0.011]
perfs2 = [p * 20 for p in perfs]
bandits = {i: Bandit(perf) for i, perf in enumerate(perfs2)}
user_visits = 100000
bandit_regret = np.zeros((user_visits, 0))
randoms = np.random.rand(user_visits)
```

```python
for i in range(user_visits):
    won = 0
    optimal_won = 0
    
    if randoms[i] < perfs[4]:
        optimal_won = 1
        
    samples = {i: b.sample() for i, b in bandits.items()}
    selected = max(samples.items(), key=itemgetter(1))[0]
    bandits[selected].update(randoms[i])
    
    
    if 300 <= i <= 1000 and not i % 50:
        for alpha, beta in ((b.alpha, b.beta) for b in bandits.values()):
            rv = stats.distributions.beta(alpha, beta)
            x = np.linspace(0, 1, 1000)
            y = np.apply_along_axis(rv.pdf, 0, x)
            plt.plot(x, y, label='%d %d' % (alpha, beta))
            plt.legend()
        plt.show()
```
