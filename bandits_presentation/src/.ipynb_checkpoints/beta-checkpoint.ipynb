Useful imports

```python
from __future__ import division # all division is created equal
import numpy as np
import scipy as sp
from scipy.stats import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for alpha, beta in [(1, 1), (3, 1), (4, 2), (3, 3), (4, 5), (2, 5)]:
    rv = stats.distributions.beta(alpha, beta)
    x = np.linspace(0, 1, 1000)
    y = np.apply_along_axis(rv.pdf, 0, x)
    plt.plot(x, y, label='%d %d' % (alpha, beta))
    plt.legend()
plt.show()
```
